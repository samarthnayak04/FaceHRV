```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv("C:/Users/sam/CODEs/t4/hrv_stress_data.csv")
print(data.head())
```

```python

plt.figure(figsize=(10, 6))
plt.scatter(data['RMSSD'], data['SDNN'], c=data['condition'], cmap='bwr', alpha=0.7)
plt.title('Scatter Plot of RMSSD vs SDNN')
plt.xlabel('RMSSD')
plt.ylabel('SDNN')
plt.colorbar(label='Condition (0: Non-Stressed, 1: Stressed)')
plt.grid()
plt.show()
```

```python
import seaborn as sns

# Create box plots for RMSSD
plt.figure(figsize=(12, 6))
sns.boxplot(x='condition', y='RMSSD', data=data)
plt.title('Box Plot of RMSSD by Condition')
plt.xlabel('Condition')
plt.ylabel('RMSSD')
plt.grid()
plt.show()

```

```python
plt.figure(figsize=(12, 6))
plt.hist(data['RMSSD'], bins=30, alpha=0.5, label='RMSSD')
plt.hist(data['SDNN'], bins=30, alpha=0.5, label='SDNN')
plt.title('Histogram of RMSSD and SDNN')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()
plt.grid()
plt.show()
```
